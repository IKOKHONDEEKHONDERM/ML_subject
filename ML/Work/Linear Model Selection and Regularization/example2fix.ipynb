```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split

# Number of samples
sample = 400

# Generate noise
sample_data = np.random.randn(sample) * 0.1  # Adjust the noise level as needed

# Generate x values and sine wave y values
x = np.linspace(0, 1, sample)
y = np.sin(2 * np.pi * x)

X = x.reshape(-1, 1)
y = y.reshape(-1, 1)

# Add noise to the sine wave data points
y_sample_data = y + sample_data.reshape(-1, 1)

plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
plt.plot(x, y, '-g')
plt.scatter(X, y_sample_data, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Wave with Noisy Data')
plt.grid(True)



X_train_val1, X_test1, y_train_val1, y_test1 = train_test_split(X, y_sample_data, test_size=0.2, random_state=12)
X_train1, X_val1, y_train1, y_val1 = train_test_split(X_train_val1, y_train_val1, test_size=0.25, random_state=12)


plt.subplot(2,2,2)
plt.plot(x, y, '-g')
plt.scatter(X_train1, y_train1, color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Wave with Noisy Data')
plt.grid(True)

```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import make_pipeline

def PolynomialRegression(degree, lambd,**kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                         Ridge(alpha=lambd))

X_range1 = np.linspace(X_val1.min(), X_val1.max(), len(X_val1))[:, np.newaxis]
X_range_test1 = np.linspace(X_test1.min(), X_test1.max(), len(X_test1))[:, np.newaxis]

X_train1.reshape(-1, 1)
y_train1.reshape(-1, 1)

model_poly1 = PolynomialRegression(60,0) 
model_poly2 = PolynomialRegression(3,10)
model_poly3 = PolynomialRegression(60,10)

model_poly1.fit(X_train1, y_train1)
model_poly2.fit(X_train1, y_train1)
model_poly3.fit(X_train1, y_train1)

y_pred1 = model_poly1.predict(X_range1)
y_pred2 = model_poly2.predict(X_range1)
y_pred3 = model_poly3.predict(X_range1)

y_pred_test1 = model_poly1.predict(X_range_test1)
y_pred_test2 = model_poly2.predict(X_range_test1)
y_pred_test3 = model_poly3.predict(X_range_test1)
```

```python
# degree  =  =  sample  n -->  overfit  lambda  =      overfit
plt.figure(figsize=(15,8))

plt.subplot(2,3,1)
plt.plot(x, y, '-g')
plt.plot(X_range1, y_pred1, '-r', label='Degree 50, lambda 0')
plt.scatter(X_val1, y_val1, color='blue', label='Sample data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Train')
plt.legend()
plt.grid(True)

plt.subplot(2,3,2)
plt.plot(x, y, '-g')
plt.plot(X_range1, y_pred2, '-r', label='Degree 3, lambda 0.3')
plt.scatter(X_val1, y_val1, color='blue', label='Sample data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Train')
plt.legend()
plt.grid(True)

plt.subplot(2,3,3)
plt.plot(x, y, '-g')
plt.plot(X_range1, y_pred3, '-r', label='Degree 50, lambda 0.3')
plt.scatter(X_val1, y_val1, color='blue', label='Sample data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Train')
plt.legend()
plt.grid(True)

plt.subplot(2,3,4)
plt.plot(x, y, '-g')
plt.plot(X_range_test1, y_pred_test1, '-r', label='Degree 50, lambda 0')
plt.scatter(X_test1, y_test1, color='blue', label='Sample data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Test')
plt.legend()
plt.grid(True)

plt.subplot(2,3,5)
plt.plot(x, y, '-g')
plt.plot(X_range_test1, y_pred_test2, '-r', label='Degree 3, lambda 0.3')
plt.scatter(X_test1, y_test1, color='blue', label='Sample data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Test')
plt.legend()
plt.grid(True)

plt.subplot(2,3,6)
plt.plot(x, y, '-g')
plt.plot(X_range_test1, y_pred_test3, '-r', label='Degree 50, lambda 0.3')
plt.scatter(X_test1, y_test1, color='blue', label='Sample data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Test')
plt.legend()
plt.grid(True)



```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score,KFold


train_score1 = []
test_score1 = []

train_score2 = []
test_score2 = []

lamb1 = 15
lamb2 = 15

def Validation_Curve(degree, lambd, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
    kf = KFold(n_splits=10, random_state=1, shuffle=True)
    model = PolynomialRegression(degree, lambd)

    test_score = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=kf)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    train_score = mean_squared_error(y_test, y_pred)
    return train_score, np.mean(-test_score)

for i in range(lamb1):
    output = Validation_Curve(60, i, X, y)
    train_score1.append(output[0])
    test_score1.append(output[1])

for i in range(lamb2):
    output = Validation_Curve(3, i, X, y)
    train_score2.append(output[0])
    test_score2.append(output[1])

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(range(lamb1),train_score1)
plt.plot(range(lamb1),test_score1)
plt.xlabel('Polynomail')
plt.ylabel('error')
plt.title("Degree 60")
plt.axvline(x=10, color='gray', linestyle='--')  
print(train_score1)
print(test_score1)

plt.subplot(1,2,2)
plt.plot(range(lamb2),train_score2)
plt.plot(range(lamb2),test_score2)
plt.xlabel('Polynomail')
plt.ylabel('error')
plt.title("Degree 3")
plt.axvline(x=10, color='gray', linestyle='--')  
print(train_score2)
print(test_score2)
```
